1.2 Write the corresponding rotation matrix, and compute the vector found by rotating u by the angle theta. Plot the vectors.

```python
import numpy as np
import matplotlib.pyplot as plt

# Vector u as provided
u = np.array([[1], [2]])

# Theta as provided converted to radian
theta = np.radians(30)

# Rotation matrix as per formula
R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])

# Compute the rotated vector using dot product
u_rotated = R.dot(u)

# Plotting
plt.figure(figsize=(6,6))
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='r', label='Original Vector (u)')
plt.quiver(0, 0, u_rotated[0], u_rotated[1], angles='xy', scale_units='xy', scale=1, color='b', label='Rotated Vector (u\')')
plt.xlim(-2, 2)
plt.ylim(-1, 3)
plt.axvline(x=0, color='grey', lw=1)
plt.axhline(y=0, color='grey', lw=1)
plt.grid(True)
plt.legend()
plt.title('Vector Rotation')
plt.show()
```

1.3 Using Python, determine whether the given vectors are a solution of the system

x_1 - 4x_2 + 3x_4 = 6
 
x_3 - 2x_4 = -3

```python
import numpy as np

##define arrays
arr23 = np.array([[1], [-2], [-5], [-1]])
arr24 = np.array([[2], [0], [-1], [1]])
arr25 = np.array([[3], [0], [2], [1]])
arr26 = np.array([[4], [1], [1], [2]])


#define function called is_solution which returns true/false on whether vectors are a solution to the system
def is_solution(vector):
    x1, x2, x3, x4 = vector.flatten()
    eq1 = x1 - 4*x2 + 3*x4
    eq2 = x3 - 2*x4
    return eq1 == 6 and eq2 == -3

#print out solutions
print(f"#23 is a solution: {is_solution(arr23)}")
print(f"#24 is a solution: {is_solution(arr24)}")
print(f"#25 is a solution: {is_solution(arr25)}")
print(f"#26 is a solution: {is_solution(arr26)}")
```

1.4 Using Python matrix operators find the solution of

x1 - 2x2 = -6

-2*x1 + 3*x2 = 7

```python
import numpy as np

##define arrays as equations A and B
eqA = np.array([[1,-2], [-2,3]])
eqB = np.array([[-6], [-7]])

#use numpy's linear algebra solve function
solution = np.linalg.solve(eqA, eqB)


print(solution)
```
